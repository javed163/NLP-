```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer

# Download necessary NLTK resources if you haven't already
nltk.download('punkt')
nltk.download('stopwords')

# Initialize stemmer
stemmer = PorterStemmer()

corpus = []
for paragraph in nltk.sent_tokenize(paragraphs): # Iterate over sentences
  text = re.sub('[^a-zA-Z]', ' ', paragraph) # Remove non-alphanumeric characters
  text = text.lower() # Convert to lowercase
  text = text.split()  # Split into words

  # Remove stop words and apply stemming
  stemmed_words = [stemmer.stem(word) for word in text if word not in set(stopwords.words('english'))]

  # Join stemmed words back into a string
  processed_text = ' '.join(stemmed_words)

  corpus.append(processed_text)  # Append the processed text to corpus

# Create a CountVectorizer object and fit it to the corpus
cv = CountVectorizer(binary=True, ngram_range=(2,3))
x = cv.fit_transform(corpus).toarray()

# Print the vocabulary
print(cv.vocabulary_)
```
